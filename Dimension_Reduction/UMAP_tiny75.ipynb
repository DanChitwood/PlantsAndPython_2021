```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import umap
#pip install plotly
import plotly.express as px
#pip install kaleido
```

```python
dataraw = pd.read_csv("MapRateFiltered_tiny_75.csv")
dataraw.head()
```

```python
datafilter= [col for col in dataraw if col.startswith('AT')]
dataquant = dataraw[datafilter]
dataquant.head()
```

```python
UMAP_2d = umap.UMAP(n_neighbors=20,min_dist=.1,n_components=2)
UMAP_3d = umap.UMAP(n_neighbors=20,min_dist=.1,n_components=3)
proj_2d = UMAP_2d.fit_transform(dataquant)
proj_3d = UMAP_3d.fit_transform(dataquant)
print(proj_2d.shape)
print(proj_3d.shape)
```

```python
fig_2d = px.scatter(
    proj_2d, x=0, y=1,
    color=dataraw.Tissue, labels={'color': 'tissue type'}
)
fig_3d = px.scatter_3d(
    proj_3d, x=0, y=1, z=2,
    color=dataraw.Tissue, labels={'color': 'tissue type'}
)
fig_3d.update_traces(marker_size=3)

fig_2d.show()
fig_3d.show()
```

```python
fig_2d.write_image("UMAP_1000Gene_2D.png")
fig_3d.write_image("UMAP_1000Gene_3D.png")
fig_3d.write_html("UMAP_1000Gene.html")
```
